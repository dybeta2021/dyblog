```python
from numpy import arange, log
from wing_model import WingModel
from matplotlib import pylab

wing = WingModel()
```

```python
ref = atm = 5000
vr = 0.2
vcr = 0
sr = 0.02
scr = 0
ssr = 1
pc = 3
cc = 3
dc = -0.2
uc = 0.2
dsm = 1
usm = 1
X = log(arange(2500, 9000, 50) / 5000)

infer_iv = wing.wing(x=X, ref=ref, atm=atm, vr=vr, vcr=vcr, sr=sr, scr=scr, ssr=ssr, pc=pc, cc=cc, dc=dc, uc=uc, dsm=dsm, usm=usm)
```

```python
fig = pylab.figure(figsize=(12, 6))
ax = fig.add_subplot(111)
pylab.plot(X, infer_iv)
pylab.axvline(0.2, color="k", ls="--")
pylab.axvline(-0.2, color="k", ls="--")
pylab.axvline(0.4, color="g", ls="--")
pylab.axvline(-0.4, color="g", ls="--")
pylab.title("Wing Model")
pylab.grid()
pylab.show()



pylab.tight_layout()
pylab.savefig("WingModel.png", dpi=600)
```
